```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Generate x values (from 0 to 4π) 
x = np.linspace(0, 4 * np.pi, 1000)

# Compute the cosine of x values
y = np.cos(x)

# Plotting the cosine wave
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Cosine wave')

# Adding labels and title
plt.title('Cosine Wave')
plt.xlabel('X values (radians)')
plt.ylabel('Cosine of X')

# Show a grid
plt.grid(True)

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```
